```python
import unicodedata

#Function to remove invisible characters and normalize spaces
def clean_text(text):
    text = unicodedata.normalize('NFKD', text) #Normalize Unicode characters
    text = ''.join([c for c in text if not unicodedata.combining(c)])  #Remove diacritical marks (accents)
    text = text.replace('-', ' ') #Replace dashes with spaces
    text = ' '.join(text.split()) #Remove multiple spaces
    return text.lower().strip()

#Apply the cleaning function to the entire column
df['Color'] = df['Color'].apply(clean_text)

#Final grouping
df['Color'].replace({
    'yello green': 'yellow green',  #Consolidate possible final variations
    'yellowgreen': 'yellow green',  #Ensure that all 'yellowgreen' are replaced
}, inplace=True)

#Check the groupings
df['Color'].value_counts()
```

```python
import plotly.express as px

coluna = df['Acidity']

#Calculate the number of classes and range using Sturges' Rule
n = len(coluna)
num_classes = int(1 + 3.322 * np.log10(n))
min_val = coluna.min()
max_val = coluna.max()
amplitude = (max_val - min_val) / num_classes

print("Número de classes:", num_classes)
print("Amplitude das classes:", amplitude)

# Plotting the interactive histogram with Plotly Express
fig = px.histogram(df, x='Acidity', nbins=num_classes, title='Frequência de notas de Acidez')

#Customizing the chart
fig.update_layout(
    xaxis_title='Valores',
    yaxis_title='Frequência',
    bargap=0.2,
    template='seaborn'
)


fig.show()

```
